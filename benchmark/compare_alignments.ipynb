## Ground Truth

```python
xyz = "partially_unfolded_beta_sheet.xyz"
traj = "partially_unfolded_beta_sheet.xyz"
ground_truth_coord = md.Universe(xyz,traj)
ground_truth_avg = ground_truth_coord.atoms.positions-ground_truth_coord.atoms.center_of_geometry()
```

```python
# make k matrix
kMat = np.zeros((ground_truth_coord.atoms.n_atoms,ground_truth_coord.atoms.n_atoms),dtype=np.float64)
from MDAnalysis.analysis.distances import self_distance_array
distances = self_distance_array(ground_truth_avg)
thresh = 0.01
otherCount = 3
i = 0
for atom1 in range(ground_truth_coord.atoms.n_atoms-1):
    for atom2 in range(atom1+1,ground_truth_coord.atoms.n_atoms):
        dist = distances[i]
        if np.abs(dist-1.0) < thresh:
            k = 100.0
            x0 = 1.0
            bi = 1
        elif np.abs(dist-1.7) < thresh:
            k = 10.0
            x0 = 1.7
            bi = 2
        else:
            k = np.round(7.5*np.exp(-np.abs(dist-2.0)),2)
            x0 = dist
            bi = otherCount
            otherCount += 1
        kMat[atom1,atom2] = k
        kMat[atom2,atom1] = k
        i += 1
# finish kmat by adding diagonals
for atom1 in range(ground_truth_coord.atoms.n_atoms):
    for atom2 in range(ground_truth_coord.atoms.n_atoms):
        if (atom1 != atom2):
            kMat[atom1,atom1] -= kMat[atom2,atom1]
```

```python
ground_truth_precision = -kMat / (310*1.9872e-3)
```

```python
# make Hessian
ground_truth_hess = np.zeros((ground_truth_coord.atoms.n_atoms*3,ground_truth_coord.atoms.n_atoms*3),dtype=np.float64)
for atom1 in range(ground_truth_coord.atoms.n_atoms):
    for atom2 in range(ground_truth_coord.atoms.n_atoms):
        if atom1 != atom2:
            #sepVec = ground.atoms.positions[atom1] - ground.atoms.positions[atom2]
            sepVec = ground_truth_avg[atom1] - ground_truth_avg[atom2]
            sepVec /= np.linalg.norm(sepVec)
            ground_truth_hess[atom1*3:atom1*3+3,atom2*3:atom2*3+3] = -2.0*kMat[atom1,atom2]*np.outer(sepVec,sepVec)
# finish Hessian by populating diagonal elements
for atom1 in range(ground_truth_coord.atoms.n_atoms):
    for atom2 in range(ground_truth_coord.atoms.n_atoms):
        if (atom1 != atom2):
            ground_truth_hess[atom1*3:atom1*3+3,atom1*3:atom1*3+3] -= ground_truth_hess[atom2*3:atom2*3+3,atom1*3:atom1*3+3]
# compute covar
ground_truth_covar = np.linalg.pinv(ground_truth_hess/(310*1.9872e-3),rcond=1e-10)
```

# Uniform alignment

# Plots

```python
def norm_covar(covar):
    N = covar.shape[0]//3
    covar_norm = np.empty((N,N),dtype=np.float64)
    for i in range(N):
        for j in range(N):
            covar_norm[i,j] = np.linalg.norm(covar[i*3:i*3+3,j*3:j*3+3])
    return covar_norm

def trace_covar(covar):
    N = covar.shape[0]//3
    covar_trace = np.empty((N,N),dtype=np.float64)
    for i in range(N):
        for j in range(N):
            covar_trace[i,j] = np.trace(covar[i*3:i*3+3,j*3:j*3+3])
    return covar_trace
```

```python
ground_truth_covar_norm = norm_covar(ground_truth_covar)
ground_truth_covar_trace = trace_covar(ground_truth_covar)
uniform_covar_norm = norm_covar(uniform_aligned_covar)
uniform_covar_trace = trace_covar(uniform_aligned_covar)
intermediate_covar_norm = norm_covar(intermediate_aligned_covar)
intermediate_covar_trace = trace_covar(intermediate_aligned_covar)
weighted_covar_norm = norm_covar(weighted_aligned_covar)
weighted_covar_trace = trace_covar(weighted_aligned_covar)
```

```python
# compute eigenvalues
e_gt = np.linalg.eigh(ground_truth_covar)[0]
e_u = np.linalg.eigh(uniform_aligned_covar)[0]
e_i = np.linalg.eigh(intermediate_aligned_covar)[0]
e_w = np.linalg.eigh(weighted_aligned_covar)[0]
# setup figure
fontsize=16
fig = plt.figure(figsize=(6,6),dpi=80)
ax = plt.subplot(111) 
xlabel="Index"
ylabel="Covariance Eigenvalue ($\AA^2$)"
ax.set_xlabel(xlabel,fontsize=fontsize)
ax.set_ylabel(ylabel,fontsize=fontsize)
ax.grid(b=True, which='major', axis='both', color='#808080', linestyle='--')
plt.tick_params(axis='both',labelsize=fontsize)
# make plots
ax.plot(e_u,'-o',lw=3,label="Uniform")
ax.plot(e_i,'-o',lw=3,label="Intermediate")
ax.plot(e_w,'-o',lw=3,label="Weighted")
ax.plot(e_gt,'-o',lw=3,label="Ground Truth")
plt.legend(fontsize=fontsize)
```

# Compare $N x N$ Covars

```python
import matplotlib.cm as cm
def plot_multiple_sq_mat(sq_data, labels, plot_layout, cbar_label, cmap=plt.cm.bwr, figsize=(10,10),dpi=80,fontsize=16):
    assert(sq_data.shape[2]==plot_layout[0]*plot_layout[1])
    plotting_cmap = cmap
    #plotting_cmap.set_under(color='white')
    numRes = sq_data.shape[0]
    delta = numRes//10
    print(delta)
    tick_labels = list(range(1,(numRes+1)//delta))
    tick_labels = [str(x) for x in tick_labels]
    tick_locations = np.arange(delta/2,numRes+delta/2,delta) 
    nNodes = sq_data.shape[0]
    node_range = range(nNodes+1)
    
    vmin = np.amin(sq_data)
    vmax = np.amax(sq_data)
    vmin = min(vmin,-vmax)
    vmax = max(-vmin,vmax)
    
    
    fig, axs = plt.subplots(plot_layout[0], plot_layout[1],figsize=figsize,dpi=dpi,sharex=True,sharey=True)
    count = 0
    for i in range(plot_layout[0]):
        for j in range(plot_layout[1]):
#    for ax in axs:
            axs[i,j].tick_params(which='major',length=6,width=2)
            axs[i,j].tick_params(which='minor',length=3,width=1)

            temp = axs[i,j].pcolormesh(node_range,node_range,sq_data[:,:,count],cmap=plotting_cmap,vmin=vmin,vmax=vmax)

            temp.axes.set_yticks(tick_locations)
            temp.axes.set_yticklabels(tick_labels)
            temp.axes.set_xticks(tick_locations)
            temp.axes.set_xticklabels(tick_labels)
            ## segnaming beside/underneath axes
            if j == 0:
                axs[i,j].set_ylabel('Residue Index',size=fontsize,labelpad=5)
            if i == plot_layout[0] - 1:
                ax.set_xlabel('Residue Index',size=fontsize,labelpad=5)
            axs[i,j].set_xlim((-0.5,nNodes+0.5))
            axs[i,j].set_ylim((-0.5,nNodes+0.5))
            axs[i,j].set_title(labels[count],fontsize=fontsize)
            axs[i,j].set_aspect('equal')
            count += 1
#    cb1 = plt.colorbar()
#    cb1.set_label(r'%s'%(cbar_label))
    plt.tight_layout()
```

```python
sq_data = np.empty((12,12,4))
sq_data[:,:,0] = ground_truth_covar_trace
sq_data[:,:,1] = uniform_covar_trace
sq_data[:,:,2] = intermediate_covar_trace
sq_data[:,:,3] = weighted_covar_trace
labels = ["Ground Truth", "Uniform", "Intermediate", "Weighted" ]
plot_multiple_sq_mat(sq_data,labels, (2,2), "Trace Covariance")
```
